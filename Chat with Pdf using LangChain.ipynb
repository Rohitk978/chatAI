```python
from PyPDF2 import PdfReader
import openai
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
```

```python
import os
os.environ["OPEN_API_KEY"] = "Your API KEY"
```

```python
pr = PdfReader('COA_Full.pdf')
```

This library is used for reading every page from the document.

```python
from typing_extensions import Concatenate

raw_text  = ' '

for i,page in enumerate(pr.pages):
  content = page.extract_text()
  if content:
    raw_text+=content
```

```python
text_splitter =  CharacterTextSplitter(
    separator = "\n",
    chunk_size = 800,
    chunk_overlap = 200,
    length_function = len
)

texts = text_splitter.split_text(raw_text)
```

```python
embed = OpenAIEmbeddings()
```

```python
document_search = FAISS.from_texts(texts,embed)
```

```python
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
```

```python
chain = load_qa_chain(OpenAI(),chain_type='stuff')
```

```python
query = "explain Instruction cycle  ?"
docs = document_search.similarity_search(query)
```

```python
chain.run(input_documents = docs, question = query)
```
